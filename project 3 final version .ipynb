```python
# read the file
def name(e):
    return e[4]

def date(e):
    return e[0]

def time(e):
    return e[1]

fh=open('userlog.log')
print(fh.read)
raw_data=fh.readlines()
arr = []

for i in raw_data:
    i=i.split()
    arr.append(i)
arr = sorted(arr, key = lambda x: (name(x), date(x), time(x)))


activity_dict = {}
people_dict = {}
same_name = []
list1 = []
list2 = []

oldname = arr[0][4]
for ele in arr:
    list1 = []
    activity = ele[2]
    name = ele[4]
    day = ele[0]
    time = ele[1]
    ID = ele[3]

    if name == oldname:
        list1.append(time)
        list1.append(activity)
        list1.append(ID)
        list_activity = activity_dict.get(day, [])
        list_activity.append(list1)
        activity_dict.update({day: list_activity})
    else:
        people_dict.update({oldname: activity_dict})
        activity_dict = {}
        oldname = name
print(" ")

```

```python
#Domain Count 
def name(e):
    return e[4]

def date(e):
    return e[0]

def time(e):
    return e[1]

fh=open('userlog.log')

raw_data=fh.readlines()
arr = []

for i in raw_data:
    i=i.split()
    arr.append(i)
   
arr = sorted(arr, key = lambda x: (name(x), date(x), time(x)))


email=[]
people_name=arr[0][4]
for y in range (0,len(arr),1):
    people_name=arr[y][4]
    email.append(people_name)
print(len(set(email)))
setmail=set(email)
print(setmail)

# how many different IDs (domains)
#IDC(setmail) is the function for counting domains
distinct_email=list(setmail)
string_email="@".join(distinct_email)
    #print(string_email)
distinct_email=string_email.split("@") 
ID=[]
for i in range (1, len(distinct_email), 2):
    ID.append(distinct_email[i])
print(ID)
count_ID={}
for ele in ID:
    if ele in count_ID:
        count_ID[ele]+=1
    else:
        count_ID[ele]=1




f=open("domain_report1.txt",'w')
f.write("DOMAIN COUNT"+"\n")
for i in count_ID:
    f.write(str(i)+"\t"+'\t'+str(count_ID[i])+"\n")
f.close()
```
